```python
#Coaches 2008

import requests, pandas as pd
from bs4 import BeautifulSoup as bs

url = 'https://www.sports-reference.com/cfb/years/2008-coaches.html'
year=2008

r=requests.get(url)
soup=bs(r.content)
table = soup.select('table.sortable stats_table now_sortable is_sorted.tr data-row')
table=soup.find_all('table')
df1=pd.read_html(str(table))[0]

del df1['Rk']
del df1['Notes']

df1['Year'] = year
```
