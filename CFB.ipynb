### Manipulating webscrapes for cfb

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
```

```python
df = pd.read_excel('U:\Administrative\Personal\cfb\cfb_rank.xlsx')
df=df[df['Rank'] < 165]
df=df[df['Rank'] > 0]
df=df[df['Year'] < 2023]
df=df.sort_values(by='Year')
#df=df.set_index('Year')
df
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_NCAA_Division_I_FBS_football_programs#:~:text=FBS%20programs%20%20%20%20Team%20%20,%20NSC%2C%20SoCon%20%2021%20more%20rows%20'
r=requests.get(url)
soup=bs(r.content)

table = soup.select('table.wikitable.sortable.jquery-tablesorter')
table=soup.find_all('table')
df1=pd.read_html(str(table))[0]

del df1['Formerconferences']
del df1['JoinedFBS']
df1.columns = ["Team","Nickname","City",'State','Enrollment','Conference','First CFB Year']
df1
#table = soup.select('')
#table = soup.find_all('table')
#df = pd.read_html(str(table))[0]
```

```python
df=df.merge(df1, on= 'Team')   #This merge parameter doesn't include all of the recruiting data, but it grabs the biggest ones that we would want to analyze. 
df
```
